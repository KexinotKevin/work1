```python
import pandas as pd 
from datasets_cfg import *
```

```python
abcd_paths = get_dataset_cfg("ABCD")
abcd_demo_table=pd.read_csv(abcd_paths["demographics_path"])
abcd_scores_table=pd.read_csv(abcd_paths["scores_path"])
```

```python
abcd_demo_table=abcd_demo_table[abcd_demo_table["eventname"]=="baseline_year_1_arm_1"]
abcd_scores_table=abcd_scores_table[abcd_scores_table["eventname"]=="baseline_year_1_arm_1"]
```

```python
tgt_scores_labels=abcd_paths["tgt_label_list"][3:]
```

```python
# 定义需要提取的列
demo_cols = ['src_subject_id', 'eventname', 'demo_brthdat_v2', 'demo_sex_v2']
```

```python
# 确保 scores 列存在
available_score_cols = [col for col in tgt_scores_labels if col in abcd_scores_table.columns]
print(f"将提取的分数列: {available_score_cols}")

# 从 scores 表中选择对齐列 + 目标分数列
scores_cols = ['src_subject_id', 'eventname'] + available_score_cols
scores_subset = abcd_scores_table[scores_cols]

# 从 demo 表中选择指定列
demo_subset = abcd_demo_table[demo_cols]

# 合并两个表（按 src_subject_id 和 eventname 对齐）
merged_df = pd.merge(
    demo_subset,
    scores_subset,
    on=['src_subject_id', 'eventname'],
    how='inner'  # 只保留两边都有的记录
)

# 查看结果
print(f"合并后数据形状: {merged_df.shape}")
print(f"列名: {merged_df.columns.tolist()}")
abcd_csv_path="/media/shulab/WD_10T/datasets/ABCD/textfiles/"
# 保存为 CSV
merged_df.to_csv(f'{abcd_csv_path}abcd_baseline_rearranged_scores.csv', index=False)
print("已保存到 abcd_baseline_rearranged_scores.csv")
```
